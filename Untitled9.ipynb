```python
from rembg import remove
# from rembg.bg import remove
from PIL import Image
import os

image_path = "/content/drive/MyDrive/drone_0618/"
image_files = [f for f in os.listdir(image_path)]

for file_name in image_files:
    output_folder = os.path.join("/content/drive/MyDrive/drone_removebg/", file_name)
    img_path = os.path.join(image_path, file_name)
    image = Image.open(img_path).convert("RGBA")
    out = remove(image)

    width, height = out.size

    output_path = os.path.join(output_folder)
    print(output_path)
    out.save(output_path)
```
