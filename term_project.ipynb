```python
import tensorflow as tf
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Conv2D, MaxPooling2D
from tensorflow.keras.optimizers import Adam
import os
from PIL import Image
import numpy as np
from sklearn.model_selection import train_test_split
from tensorflow.keras.preprocessing.image import ImageDataGenerator

images = []
Y = []
n = 0
dataset_path = "/content/drive/MyDrive/dog"
# class_folders = [folder for folder in os.listdir(dataset_path) if os.path.isdir(os.path.join(dataset_path, folder))]

for folder_name in os.listdir(dataset_path):
    folder_dir = os.path.join(dataset_path, folder_name)

    for file_name in os.listdir(folder_dir):
            img_path = os.path.join(folder_dir, file_name)
            image = Image.open(img_path).resize((256,256))
            images.append(image)
            Y.append(n)

    n += 1

X = np.stack(images)
Y = to_categorical(Y)
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# 이미지 데이터 증강을 위한 ImageDataGenerator 생성
datagen = ImageDataGenerator(
    rotation_range=20,  # 이미지 회전 범위
    width_shift_range=0.1,  # 가로 방향으로 이동 범위
    height_shift_range=0.1,  # 세로 방향으로 이동 범위
    shear_range=0.2,  # 전단 강도 범위
    zoom_range=0.2,  # 확대/축소 범위
    horizontal_flip=True,  # 수평 뒤집기
    fill_mode='nearest'  # 새로 생성된 픽셀 채우기 방식
)

# 데이터 증강을 적용하여 추가 학습 데이터 생성
augmented_images = []
augmented_labels = []

for image, label in zip(X_train, Y_train):
    # 현재 이미지를 배치 차원으로 변경 (ImageDataGenerator.flow() 함수의 요구사항)
    image = np.expand_dims(image, axis=0)
    
    # 데이터 증강 수행
    augmented_images.extend(datagen.flow(image, batch_size=1)[0])
    augmented_labels.extend([label] * 1)  # 증강된 이미지 수에 맞게 레이블 복제

# 증강된 이미지를 기존 학습 데이터에 추가
X_train_augmented = np.concatenate([X_train, augmented_images])
Y_train_augmented = np.concatenate([Y_train, augmented_labels])


model = Sequential()
model.add(Conv2D(32,(3,3), input_shape=(256,256,3), activation='relu', padding='same'))
model.add(MaxPooling2D(2,2))

model.add(Conv2D(64, (3,3), activation='relu'))
model.add(MaxPooling2D(2,2))

model.add(Conv2D(128, (3,3), activation='relu'))
model.add(MaxPooling2D(2,2))

model.add(Conv2D(256, (3,3), activation='relu'))
model.add(MaxPooling2D(2,2))

model.add(Conv2D(512, (3,3), activation='relu'))
model.add(MaxPooling2D(2,2))

model.add(Conv2D(1024, (3,3), activation='relu'))
model.add(MaxPooling2D(2,2))

model.add(Flatten())
model.add(Dense(1024, activation='relu'))
# model.add(Dense(128, activation='relu'))
model.add(Dense(5, activation='softmax'))

model.compile(loss='categorical_crossentropy',optimizer=Adam(learning_rate=5E-04), metrics=['accuracy'])

history = model.fit(X_train, Y_train, epochs=100, batch_size=16, validation_data=(X_test, Y_test))
#data augmentation
# history = model.fit(X_train_augmented, Y_train_augmented, epochs=100, batch_size=16, validation_data=(X_test, Y_test))

print("Test loss : %.4f" % (model.evaluate(X_test, Y_test)[0]))
print("Test accuracy : %.4f" %(model.evaluate(X_test, Y_test)[1]))
```
